```python
import openai
```

```python
openai.File.create(file=open('classify.jsonl'), purpose='classifications')
```

```python
_.id
```

```python
(file_id:=_.id)
```

```python
file_id
```

```python
openai.Classification.create(
    file=file_id,
    model='curie-msft',
    search_model='ada-msft',
    max_examples=3,
    query='the bainghan bhartha was very spicy.  the tandoori chicken was cooked perfectly and the naan is really fantastic'

)
```

```python
openai.File.create(file=open('classify.jsonl'), purpose='search')
```

```python
(search_file:=_.id)
```

```python
openai.Engine('ada-msft').search(
    search_model='ada-msft',
    query='hummus',
    max_rerank=5,
    file=search_file
)
```

```python
openai.File.create(file=open('classify.jsonl'), purpose='answers')
```

```python
(qa_file:=_.id)
```

```python
openai.Answer.create(
    search_model='ada-msft',
    model='curie-msft',
    question='What restaurant has good sushi and tempura?',
    file=qa_file,
    max_rerank=10,
    max_tokens=5,
    stop=["\n"],
    examples=[
        ['What is the best restaurant in the city?', 'Canlis is a really good restaurant']
    ],
    examples_context='Restaurants are really fun places to eat'
)
```
